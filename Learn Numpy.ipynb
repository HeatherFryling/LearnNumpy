```python
import numpy as np
# numpy is always imported as np
```

```python
# Create an array with three zeros.
# Type is float.
a = np.zeros(3)
a
```

```python
# Create a bigger array with zeros.
z = np.zeros(10)
z
```

```python
# Examining the shape.
# 1d
z.shape
```

```python
# Changing to 2d.
# Array is now 10 rows, 1 column.
z.shape = (10, 1)
z
```

```python
# Can also create ones.
# Data type is also float.
z = np.ones(10)
z
```

```python
# Using linspace to create divisions within an inclusive range.
# These might, for example, be used as tick marks in a graph.
z = np.linspace(2, 10, 5)
z
```

```python
# Create an array from a list.
z = np.array([10, 20])
z
```

```python
# Create a random array.
np.random.seed(0)
z1 = np.random.randint(10, size = 6)
z1
```

```python
# Getting an example photo to play with.
# This photo is subject to a creative commons license.
# It was created by Vladimir Tro. https://www.flickr.com/photos/149732725@N08/50299885853
from skimage import io
photo = io.imread('https://live.staticflickr.com/65535/50299885853_b9bb4861cf_k.jpg')
# The photo is imported as a numpy array.
type(photo)
```

```python
# The photo is 2048 rows, 2048 columns, and three channels (RGB).
photo.shape
```

```python
# Examining a row to see the RGB values.
photo[0]
```

```python
# Show the photo with matplotlib.
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
# Reverse the rows in the image.
plt.imshow(photo[::-1])
```

```python
# Leave the rows alone, reverse the columns in the image.
plt.imshow(photo[:, ::-1])
```

```python
# Cut out the cat's face.
# Slice the array by [row:row, col:col]
plt.imshow(photo[185:1250, 600:1550])
```

```python
# Cut the image size in half by taking every other row and every other column.
plt.imshow(photo[::2, ::2])
```

```python
# Mathematical operations on numpy arrays.
# Useless for a photo (?), but it shows you can take the sin of every value in an array with a simple function.
photo_sin = np.sin(photo)
photo_sin
```

```python
# Other mathematical operations.
print(np.sum(photo))
print(np.prod(photo))
print(np.mean(photo))
print(np.std(photo))
print(np.var(photo))
print(np.min(photo))
print(np.max(photo))
# Index of minimum element.
print(np.argmin(photo))
# Index of maximum element.
print(np.argmax(photo))
```

```python
# A small array to test functions that create arrays of booleans.
z = np.array([1, 2, 3, 4, 5])
```

```python
less = z < 3
less
```

```python
greater = z > 3
greater
```

```python
# Can use these booleans as a mask for taking sections.
values_over3 = z[z > 3]
values_over3
```

```python
# Demo this by making brights max bright
# and darks max dark.
photo_masked = np.where(photo > 100, 255, 0)
```

```python
plt.imshow(photo_masked)
```

```python
# Array addition, element by element (not concatenation).
# Arrays must be the same shape.
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 7, 8, 9, 10])
arr3 = arr1 + arr2
arr3
```

```python
# Can also add a constant to every value.
arr1 + 30
```

```python
# Multiply element by element.
arr1 * arr2
```

```python
# Multiply by a scalar.
arr1 * 10
```

```python
# Dot product!
arr1 @ arr2
```

```python
# Transpose.
plt.imshow(photo[:, :, 0].T)
```

```python
# Sort
x = np.array([2, 3, 1, 4, 5])
np.sort(x)
```
